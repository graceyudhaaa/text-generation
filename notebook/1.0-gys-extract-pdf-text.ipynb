```python
from PyPDF2 import PdfReader
```

```python
reader = PdfReader("../data/raw/Kalarahu Kumpulan Cerita Rakyat Jawa.pdf")
all_text = ""

for page in reader.pages[7:]:
    text = page.extract_text()
    text = text.replace('\xad\n', '')[3:]
    if text[0].isnumeric() and text[0] != '1':
        text = "\n" + text
    
    all_text += text
    # print(text)

print(all_text)
```

```python
with open('../data/processed/text-preprocessed.txt', 'w') as file:
    file.write(all_text)
```
